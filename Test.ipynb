```python
import numpy as np
import pandas as pd
df =pd.read_csv('titanic.csv')
df.head()
```

```python
details = {'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
           'salary': [50000, 60000, 55000, 70000, 65000],
           'exp': [2.4, 4.3, 5.5, 10, 8]}
d1 = pd.DataFrame(details)
d1
```

```python
# Filtering data: only rows where salary > 55000
d2 = d1[d1['salary'] > 55000]
d2
```

```python
# Dropping 'salary' column and displaying the result
d3 = d1.drop(['salary'], axis='columns')
d3
```

```python
# Filling null values in 'Age' with the mean age
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Verify if nulls are removed
df.isnull().sum()
```

```python
df.shape
```

```python
df1 =df.dropna(axis='rows', subset = ['Embarked'])
```

```python
# Fill Embarked nulls with the most common value (mode)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Check again
df.isnull().sum()
```

```python
print (df['Age'])
```

```python
import matplotlib.pyplot as plt

# Using 'df' (the main dataframe) and 'Age' (with capital 'A' as in your dataset)
df['Age'].hist(bins=20, edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate mean and median for Age
age_mean = df['Age'].mean()
age_median = df['Age'].median()

print(f"Mean Age: {age_mean}")
print(f"Median Age: {age_median}")

# Create a box plot for Age
plt.figure(figsize=(6, 8))
plt.boxplot(df['Age'])
plt.title('Vertical Box Plot of Passenger Ages')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
# Group by 'Sex' and calculate the mean for numeric columns (like Age, Fare)
print("Mean values by Sex:")
print(df.groupby('Sex')[['Age', 'Fare', 'Survived']].mean())

# Group by 'Pclass' and calculate the mean Age
print("\nMean Age by Pclass:")
print(df.groupby('Pclass')['Age'].mean())
```
