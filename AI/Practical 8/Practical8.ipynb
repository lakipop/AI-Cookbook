## Propositional Logic in Python 
### Task 01 – Implement Negation (NOT)

```python
def negation(p): 
    return not p 

print("P   NOT P") 

for p in [True, False]: 
    result = negation(p) 
    print(p, result)
```

### Task 02 – Implement Conjunction (AND)

```python
def conjunction(p, q): 
    return p and q 

print("P   Q   P AND Q") 

for p in [True, False]: 
    for q in [True, False]: 
        result = conjunction(p, q) 
        print(p, q, result) 
```

### Task 03 – Implement Disjunction (OR) 

```python
def disjunction(p, q): 
    return p or q 

print("P   Q   P OR Q") 

for p in [True, False]: 
    for q in [True, False]: 
        result = disjunction(p, q) 
        print(p, q, result)
```

### Task 04 – Implement Exclusive OR (XOR)

```python
def exclusive_disjunction(p, q): 
    return (p and not q) or (not p and q) 

print("P   Q   P XOR Q") 

for p in [True, False]: 
    for q in [True, False]: 
        result = exclusive_disjunction(p, q) 
        print(p, q, result)
```

### Task 05 – Implement Implication 

```python
def implication(p, q): 
    return (not p) or q 

print("P   Q   P -> Q") 

for p in [True, False]: 
    for q in [True, False]: 
        result = implication(p, q) 
        print(p, q, result)
```

### Task 06 – Implement Biconditional 

```python
def biconditional(p, q): 
    return p == q 

print("P   Q   P <-> Q") 

for p in [True, False]: 
    for q in [True, False]: 
        result = biconditional(p, q) 
        print(p, q, result)
```

### Task 07 – Compound Proposition 

Logical relationships between propositions (or statements, sentences, assertions)

```python
def conjunction(p, q): 
    return p and q 

def disjunction(p, q): 
    return p or q 

print("P   Q   RESULT") 

for p in [True, False]: 
    for q in [True, False]: 
        result = disjunction(conjunction(p, q), not q) 
        print(p, q, result)
```

### Task 09 – Logical Reasoning Example

```python
def conjunction(p, q): 
    return p and q 

def implication(p, q): 
    return (not p) or q 

print("P   Q   R   RESULT") 

for p in [True, False]: 
    for q in [True, False]: 
        for r in [True, False]: 
            result = implication(conjunction(p, q), r) 
            print(p, q, r, result)
```

### Task 10 – AI Knowledge Representation Example 

#### 1. Propositional Logic Solver in Python

Example Expression 

(P∧Q)→R  

```python
import itertools # Python library used for working with combinations and iterations.

# Logical functions 
def negation(p):            # NOT
    return not p 

def conjunction(p, q):      # AND
    return p and q 

def disjunction(p, q):      # OR
    return p or q 

def implication(p, q):      # IMPLICATION (P → Q = ¬P ∨ Q)
    return (not p) or q 

def biconditional(p, q):    # BICOMONDITIONAL (P == Q)
    return p == q 

# Evaluate logical expression (P AND Q) -> R 
def evaluate_expression(p, q, r): 
    return implication(conjunction(p, q), r) 

print("P\tQ\tR\tResult") 

for p, q, r in itertools.product([True, False], repeat=3): # itertools.product() generates all combinations.
    result = evaluate_expression(p, q, r) 
    print(f"{p}\t{q}\t{r}\t{result}")
```

#### 2. Automatic Truth Table Generator 

```python
# Import itertools library to generate combinations of truth values
import itertools 

# List of propositional variables used in the logical expression
variables = ['P', 'Q'] 

print("P   Q   P AND Q") 

# Generate all possible combinations of True and False
for values in itertools.product([True, False], repeat=len(variables)): 
    
    # Assign the generated values to P and Q
    # Example combinations: (True,True), (True,False), (False,True), (False,False)
    P, Q = values 
    result = P and Q 
    print(f"{P}\t{Q}\t{result}") 
```

#### 3. AI Reasoning Engine using Propositional Logic

Example scenario: 

1. If it didn't rain, Harry visited Hagrid 

2. Harry visited Hagrid or Dumbledore but not both 

3. Harry visited Dumbledore 

Question: Did it rain? 

```python
import itertools 

# Logical rules 
def implication(p, q): 
    return (not p) or q 

def xor(p, q): 
    return (p and not q) or (not p and q) 

print("Rain  Hagrid  Dumbledore  KnowledgeBase") 

for rain, hagrid, dumbledore in itertools.product([True, False], repeat=3): 
    rule1 = implication(not rain, hagrid) 
    rule2 = xor(hagrid, dumbledore) 
    rule3 = dumbledore 

    knowledge_base = rule1 and rule2 and rule3 

    if knowledge_base: 
        print(rain, hagrid, dumbledore, "VALID")
```

Question 01 – Smart Home Lighting System 


Smart Home Rules 

1. If temperature is high → Fan turns on 

2. If humidity is high → Fan turns on 

3. Temperature is high 

Question: 

Will the fan turn on?

```python
# Logical implication
def implication(p, q):
    return (not p) or q

temperature_high = True
humidity_high = False
fan_on = True

rule1 = implication(temperature_high, fan_on)
rule2 = implication(humidity_high, fan_on)

knowledge_base = rule1 and rule2

if knowledge_base:
    print("Fan will turn ON")
else:
    print("Fan will NOT turn ON")
```
