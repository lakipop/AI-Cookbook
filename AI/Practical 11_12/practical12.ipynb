## Neural Networks in Machine Learning

A neural network is a machine learning model inspired by how the human brain works. It is designed to learn patterns from data by connecting simple computing units called neurons.

Structure of a Neural Network

01) Input Layer - This is where data enters.

    Example: Age = 25, Salary = 50,000

02) Hidden Layers - This is the “brain” of the network. It performs calculations. It finds hidden patterns.

03) Output Layer - This gives the final result.

    Examples: 0 or 1 (spam or not spam), Class label (cat or dog)

### Task 01: Neural Network Implementation in Python 

#### Step 1: Import Libraries

```python
import numpy as np
```

#### Step 2: Define the Dataset

```python
# Inputs (XOR truth table) 
X = np.array([[0, 0], 
[0, 1], 
[1, 0], 
[1, 1]]) 

# Output labels 
Y = np.array([[0], 
[1], 
[1], 
[0]])
```

#### Step 3: Initialize Hyperparameters

Hyperparameters are settings you choose before training the model.

They control:

how big the network is,
how fast it learns, and 
how long it trains

```python
input_neurons = 2        # 2 input features
hidden_neurons = 4       # can be tuned 
output_neurons = 1       # binary output 
learning_rate = 0.1 
epochs = 10000 
```

Input layer = 2 neurons (Because input data has 2 features) - 1 neuron for each input feature

Hidden layer = 4 neurons (These neurons learn patterns , More neurons = more learning power)

Output layer = 1 neuron (XOR output is binary → 0 or 1)

epoches - Number of times the model sees the entire dataset.

1 epoch = one full pass through all data

10000 epochs = model learns 10000 times


#### Step 4: Initialize Weights and Biases

```python
np.random.seed(1)  # reproducibility 

# Weights and biases 
wh = np.random.uniform(size=(input_neurons, hidden_neurons))  # input -> hidden 
bh = np.random.uniform(size=(1, hidden_neurons)) 

wo = np.random.uniform(size=(hidden_neurons, output_neurons))  # hidden -> output 
bo = np.random.uniform(size=(1, output_neurons))
```

wh (hidden weights) -This creates a weight matrix between input layer and hidden layer. Each input neuron connects to every hidden neuron.

    input_neurons = 2, hidden_neurons = 4

bh (hidden layer bias) - Each hidden neuron gets its own bias

wo (output weights) - Each hidden neuron connects to the output neuron:

    Each connection has a weight.

    hidden_neurons = 4, output_neurons = 1

bo (output bias) - The final output neuron also has a bias term.

#### Step 5: Activation Function

```python
def sigmoid(x): 
    return 1 / (1 + np.exp(-x)) 

def sigmoid_derivative(x): # calculates how much the sigmoid output changes
    return x * (1 - x)
```

#### Step 6: Training Loop

```python
for epoch in range(epochs): 
    # Forward Propagation 
    zh = np.dot(X, wh) + bh # calculates weighted sum for hidden layer
    ah = sigmoid(zh) # Applies sigmoid, Converts values into 0–1 range
    zo = np.dot(ah, wo) + bo # calculates weighted sum for output layer
    ao = sigmoid(zo) # prediction of the model

    # Backpropagation 
    error = Y - ao # calculates error at output layer (actual values - predicted values)
    d_output = error * sigmoid_derivative(ao) # How much output layer should change
    error_hidden = d_output.dot(wo.T) # backpropagation - which hidden neurons caused that mistake
    d_hidden = error_hidden * sigmoid_derivative(ah) # How much hidden neurons should adjust

    # Update weights and biases 
    wo += ah.T.dot(d_output) * learning_rate # Adjusts weights between hidden and output layer
    bo += np.sum(d_output, axis=0, keepdims=True) * learning_rate # Adjusts output bias
    wh += X.T.dot(d_hidden) * learning_rate # Adjusts weights from input to hidden layer
    bh += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate # Adjusts hidden layer bias

    # axis = 0 means we sum across rows (for each column), keepdims=True keeps the result as a 2D array

    # Optionally print loss 
    if epoch % 1000 == 0: 
        loss = np.mean(np.square(error)) 
        print(f'Epoch {epoch}, Loss: {loss:.4f}') # Every 1000 epochs, prints error
```

#### Step 7: Final Prediction

```python
print("\nFinal Output after Training:") 
print(np.round(ao, 3)) # Rounded output to 3 decimal places for better readability
```
