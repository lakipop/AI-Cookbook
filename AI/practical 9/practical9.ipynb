#### When is the mean misleading?

The mean becomes misleading when your dataset has:

1. Outliers (extreme values) - Very high or very low values pull the mean away from the “typical” value

2. Skewed distribution - Data is not evenly distributed (clustered on one side)

#### Activity 3: Spread of Data 

Standard deviation is the spread of the statistical data from the mean or average position. 
It is represented by the Greek symbol σ.

Variance is how far the data points are spread from the mean.
It is represented by a square of the Greek symbol sigma. 

```python
data = [99,86,87,88,111,86,103,87,94,78,77,85,86]

print("Standard Deviation:", np.std(data)) 
print("Variance:", np.var(data))
```

### PART 05: Simple Linear Regression 

Simple Linear Regression is a method used to find the relationship between two variables:

One independent variable (X)

One dependent variable (Y)

#### Activity 8: Build Regression Model 

```python
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(x, y) 
print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", r)
print("P-value:", p)
print("Standard error:", std_err)
```

Slope (m) - Shows how much Y changes when X increases by 1
            
            y=mx+c


Intercept (c) - The value of Y when X = 0

r (Correlation Coefficient) - Measures strength and direction of relationship

            −1 ≤ r ≤1

            r = 1 → perfect positive relationship (When X increases, Y also increases)
            
            r = -1 → perfect negative relationship (When X increases, Y also decreases)
            
            r = 0 → no relationship


p (p-value) - Tells you if the relationship is statistically significant

            p < 0.05 → significant (relationship is real)
            
            p > 0.05 → not significant (could be random)

#### Activity 9: Plot Regression Line 

```python
def predict(x): 
    return slope * x + intercept #( y = mx + c)

model = list(map(predict, x)) 
plt.scatter(x, y) 
plt.plot(x, model) 
plt.show() 
```

#### Activity 10: Interpret Results 

```python
print("Slope:", slope) 
print("Intercept:", intercept) 
print("R value:", r)
```


What does slope represent?  

Slope (m) = rate of change of y with respect to x. Tells how much y changes when x increases by 1


What happens if slope = 0?  

Line becomes horizontal. y does not change with x


What does R value indicate? 

R = correlation coefficient. Measures strength and direction of relationship between x and y

+1	Perfect positive correlation (x increases, y increases)
-1	Perfect negative correlation (x increases, y decreases)


#### Activity 11: Prediction

```python
print("Predicted value for x=10:", predict(10))
```

#### Activity 12: Load Dataset

```python
df = pd.read_csv("data.csv") # read the dataset from a CSV file
print(df.head()) # print first 5 rows of the dataset
```

#### Activity 13: Data Exploration 

```python
print(df.describe()) # print summary statistics of the dataset
print(df.info()) # print information about the dataset, including data types and non-null counts
```

Identify independent variables  - Volume, Weight

Identify dependent variable - CO2

Car and Model are usually categorical.

#### Activity 14: Build Model

```python
from sklearn import linear_model

X = df[['Weight', 'Volume']] 
y = df['CO2'] 
model = linear_model.LinearRegression() 
model.fit(X, y)
```

#### Activity 15: Prediction 

```python
print(model.predict([[2300, 1300]])) 
```

#### Activity 16: Interpret Coefficients 

```python
print("Coefficients:", model.coef_) 
print("Intercept:", model.intercept_)
```

Which variable has more influence?

The variable with larger absolute coefficient

weight coefficient = 0.0075, 
volume coefficient = 0.0078

Interpretation:

Volume has more influence on CO2 because 0.0078 > 0.0075

For every 1 kg increase in weight, CO2 increases by 0.0075 

For every 1 cc increase in volume, CO2 increases by 0.0078

### PART 07: Model Evaluation

#### Activity 17: Train-Test Split

```python
train = df[:int(len(df)*0.8)] # use the first 80% of the data for training
test = df[int(len(df)*0.8):] # use the remaining 20% of the data for testing
```

#### Activity 18: Evaluate Model

```python
from sklearn.metrics import r2_score

train_x = np.array(train[['Weight', 'Volume']]) 
train_y = np.array(train[['CO2']]) 

test_x = np.array(test[['Weight', 'Volume']]) 
test_y = np.array(test[['CO2']]) 

model.fit(train_x, train_y) 
predictions = model.predict(test_x) 
print("R2 Score:", r2_score(test_y, predictions)) # R2 score is a measure of how well the model fits the data. It ranges from 0 to 1, where 1 indicates a perfect fit and 0 indicates that the model does not explain any of the variance in the data.
print("MAE:", np.mean(np.abs(test_y - predictions))) # MAE is a measure of the average absolute difference between the predicted and actual values.
print("MSE:", np.mean((test_y - predictions)**2)) # MSE is a measure of the average squared difference between the predicted and actual values.
```
