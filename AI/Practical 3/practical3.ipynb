## Part 1: Python Conditional Statements

###  1.1 if Statement 

```python
number = 10 
if number > 0: 
    print('Number is positive') 
```

### 1.2 if–else Statement 

```python
number = -5 
if number > 0: 
    print('Positive number') 
else: 
    print('Negative number') 
```

### 1.3 if–elif–else Statement

```python
number = 0 
if number > 0: 
    print('Positive') 
elif number == 0: 
    print('Zero') 
else: 
    print('Negative')
```

### 1.4 Nested if Statements 

```python
number = 5 
if number >= 0: 
    if number == 0: 
        print('Number is zero') 
    else: 
        print('Number is positive') 
else: 
    print('Number is negative') 
```

## Part 2: Python Loops

### 2.1 for Loop

```python
for i in range(5): 
    print(i) 
#Iterating through a list: 
    languages = ['Swift', 'Python', 'Go'] 
    for lang in languages: 
        print(lang)
```

### 2.2 for Loop with else

```python
digits = [0, 1, 5] 
for i in digits: 
    print(i) 
else: 
    print('No items left')
```

### 2.3 while Loop

```python
i = 1 
while i <= 5: 
    print(i) 
    i += 1
    counter = 0 
    while counter < 3: 
        print('Inside loop') 
        counter += 1 
    else: 
        print('Inside else') 
```

## Part 3: Loop Control Statements

### 3.1 break Statement 

```python
for i in range(5): 
    if i == 3: 
        break 
    print(i) 
```

### 3.2 continue Statement 

```python
num = 0 
while num < 10: 
    num += 1 
    if num % 2 == 0: 
        continue 
    print(num)
```

### 3.3 pass Statement

```python
n = 10 
if n > 10: 
    pass 
print('Hello') 
```

## Part 4: Python Functions 

### 4.2 Defining and Calling a Function 

```python
def greet(): 
    print('Hello World')
     
greet() 
```

### 4.3 Functions with Arguments 

```python
def add_numbers(num1, num2): 
    print(num1 + num2) 
    
add_numbers(5, 4) 
```

### 4.4 Return Statement

```python
def find_square(num): 
    return num * num 

result = find_square(3) 
print(result) 
```

### 4.5 Default Parameters

```python
def greet(name, message='Hi'): 
    return f"{message} {name}" 

print(greet('John')) 
print(greet('John', 'Hello'))
```

### 4.6 Keyword Arguments

```python
def get_net_price(price, tax=0.07, discount=0.05): 
    return price * (1 + tax - discount) 

#pass the values using parameter name
print(get_net_price(price=100, discount=0.06))
```

### 4.7 Recursive Functions 

```python
def count_down(start): 
    print(start) 

    next_num = start - 1 
    if next_num > 0: 
        count_down(next_num)

count_down(3) 
```

## Activities

```python
def check_even(number):
    if number % 2 == 0:
        print("Even")
    else:
        print("Odd")

check_even(5)
check_even(4)
```

```python
def multiplication_table(number):
    for i in range(1, 11):
        print(f"{number} x {i} = ", number * i)

multiplication_table(5)
```

```python
def factorial(number):
    if number == 0 or number == 1:
        return 1
    else:
        return number * factorial(number - 1)
    
print(factorial(6))
```

```python
def sum_natural(n):
    if n == 1:          # base case
        return 1
    else:
        return n + sum_natural(n - 1)
    
print(sum_natural(5))
```

```python
def student_details(name, age, grade, school="High School"):
    return f"Name: {name}, Age: {age}, Grade: {grade}, School: {school}"

student_details(name="Alice", age=14, grade="9th")

```
