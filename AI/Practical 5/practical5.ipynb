#### Step 9: Inheritance

One class (Child class) to inherit properties and methods from another class (Parent class).

Purpose:
    Code reuse,
    Avoid duplication, and
    Logical hierarchy

Example:
    Person → Student

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello," , self.name)

class Student(Person): # Student class inherits from Person
    def study(self):
        print(self.name, "is studying.")

stu = Student("kamal")
stu.greet()  # Inherited method'
stu.study()  # Student's own method
```

#### Step 10: Encapsulation

Binding data and methods together inside a class.

By using private variable. 

To make private variable, double underscore (__) can be used.
    Ex: __balance

Benefits:
    Data security and 
    Controlled access

```python
class BankAccount: 
    def __init__(self): 
        self.__balance = 0  # Private attribute 

    def deposit(self, amount): 
        self.__balance += amount 

    def get_balance(self): 
        return self.__balance 

account = BankAccount() # Create a BankAccount object
account.deposit(5000) 
print("Balance:", account.get_balance()) 
```

#### Step 11: Polymorphism

Same method name, different behavior.

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal): # Dog class inherits from Animal
    def speak(self):
        print("Dog barks")

class Cat(Animal): # Cat class inherits from Animal
    def speak(self):
        print("Cat meows")

animals = [Dog(), Cat(), Animal()] # List of different animal objects

for a in animals:
    a.speak()  # Polymorphic call
```

#### Question 9 – Inheritance  

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")


class Student(Person):
    def __init__(self,name, age,student_id, grade):
        super().__init__(name, age) # Call the parent class constructor
        self.student_id = student_id
        self.grade = grade   

    def display_student(self):
        self.display() 
        print(f"Student ID: {self.student_id}")
        print(f"Grade: {self.grade} \n")

student1 = Student("Alice", 20, "S12345", "A")
student1.display_student()
```

#### Question 10 – Encapsulation  

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance # Private attribute

    def deposit(self, amount):
        self.__balance += amount
        print("Deposited Successfully!")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient funds!")
        else:
            self.__balance -= amount
            print("Withdrawal Successful!")

    def get_balance(self):
        print("Final Balance:", self.__balance)

    
account = BankAccount(1000)
account.deposit(5000)
account.withdraw(2000)
account.get_balance()
```

#### Question 11 – Polymorphism 

```python
class Shape:
    def area(self):
        pass # Abstract method, to be implemented by subclasses

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        import math
        return math.pi * self.radius ** 2

class Rectangle(Shape):        
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
    
circle = Circle(5)
rectangle = Rectangle(4, 6)
print("Area of Circle:", circle.area())
print("Area of Rectangle:", rectangle.area())
```
